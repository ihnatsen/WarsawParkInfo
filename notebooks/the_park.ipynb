```python
df = pd.read_csv(r'D:\WarsawParkInfo\qgis\park.csv')

rating = df['rating'].str.replace(',', '.').astype('float')
percentiles  = [rating.quantile(0.25), rating.quantile(0.5), rating.quantile(0.9)]

def case(x):
    if x >= rating.quantile(0.9):
        return 'A'
    elif rating.quantile(0.5) <= x < rating.quantile(0.9):
        return 'B'
    elif rating.quantile(0.25) <= x < rating.quantile(0.5):
        return 'C'
    else:
        return 'D'
        
category = df['rating'].str.replace(',', '.').astype('float').apply(case)
category.name = 'category'

df = pd.concat([df, category], axis=1)
# df[['park_id', 'WKT', 'name', 'attractions', 'area', 'rating']]
```
